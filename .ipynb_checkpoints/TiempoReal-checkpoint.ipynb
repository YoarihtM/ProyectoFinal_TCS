```python
import matplotlib
import pyaudio as pa 
import struct 
import numpy as np
import scipy.fftpack as fourier
import matplotlib.pyplot as plt
from scipy.signal import find_peaks

matplotlib.use('TkAgg')
#%matplotlib inline

MUESTRA = 1024*8                                   # Tamaño del paquete a procesar
FORMATO = pa.paInt16                               # Formato de lectura INT 16 bits
CANAL = 1
Fs = 44100                                        # Frecuencia de muestreo típica para audio

p = pa.PyAudio()

stream = p.open(                                  # Abrimos el canal de audio con los parámeteros de configuración
    format = FORMATO,
    channels = CANAL,
    rate = Fs,
    input=True,
    output=True,
    frames_per_buffer=MUESTRA
)

## Creamos una gráfica con 2 subplots y configuramos los ejes

fig, (ax,ax1) = plt.subplots(2)

x_audio = np.arange(0,MUESTRA,1)
x_fft = np.linspace(0, Fs, MUESTRA)

line, = ax.plot(x_audio, np.random.rand(MUESTRA),'g')
line_fft, = ax1.semilogx(x_fft, np.random.rand(MUESTRA), 'b')

ax.set_ylim(-1500,1500)
ax.ser_xlim = (0,MUESTRA)

Fmin = 1
Fmax = 5000
ax1.set_xlim(Fmin,Fmax)

fig.show()


F = (Fs/MUESTRA)*np.arange(0,MUESTRA//2)                 # Creamos el vector de frecuencia para encontrar la frecuencia dominante

while True:
    
    data = stream.read(MUESTRA)                         # Leemos paquetes de longitud FRAMES
    dataInt = struct.unpack(str(MUESTRA) + 'h', data)   # Convertimos los datos que se encuentran empaquetados en bytes
    
    line.set_ydata(dataInt)                            # Asignamos los datos a la curva de la variación temporal
    
    M_gk = abs(fourier.fft(dataInt)/MUESTRA)            # Calculamos la FFT y la Magnitud de la FFT del paqute de datos

    
    ax1.set_ylim(0,np.max(M_gk+10)) 
    line_fft.set_ydata(M_gk)                           # Asigmanos la Magnitud de la FFT a la curva del espectro 
    prom = np.mean(M_gk)
    M_gk = M_gk[0:MUESTRA//2]                          # Tomamos la mitad del espectro para encontrar la Frecuencia Dominante
    
    peaks, _ = find_peaks(M_gk, distance=44100)
    
    print(prom)
    Posm = np.where(M_gk == np.max(M_gk))
    F_fund = F[Posm]  
    # Encontramos la frecuencia que corresponde con el máximo de M_gk
    
    
    
    if (F_fund > 150 and F_fund < 350) and (prom < .15) :                   # Rango de frecuencias para U
        print("La nota es U, con frecuencia: ",F_fund)
    elif F_fund > 150 and F_fund < 316:                   # Rango de frecuencias para I
        print("La letra es I, con frecuencia: ",F_fund)
    elif (F_fund > 200 and F_fund < 470) and (prom < 1.) :                   # Rango de frecuencias para O y prom de muestras
        print("La nota es O, con frecuencia: ",F_fund)
    
    elif F_fund > 317 and F_fund < 549:                   # Rango de frecuencias para E
        print("La letra es E, con frecuencia: ",F_fund)
    
    elif F_fund > 550 and F_fund < 1000:                     # Rango de frecuencias para 
        print("La letra es A, con frecuencia: ",F_fund)
    
    #if F_fund > 320 and F_fund < 340:                   # Rango de frecuencias para nota MI
        #print("La nota es MI, con frecuencia: ",F_fund)
        
    print(int(F_fund))                                 # Imprimimos el valor de la frecuencia dominante

    fig.canvas.draw()
    fig.canvas.flush_events()
```
